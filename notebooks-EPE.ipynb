## Antecedentes

- Inició en marzo del 2001 ante una necesidad de proveer información mensual de la situación del mercado laboral ya que en el año 2000 se había dejado de ejecutar encuestas trimestrales por razones presupuestarias. 
- Un equipo liderado por el Ministerio de Trabajo y Promoción del Empleo (MTPE), el MEF, el INEI y otros investigadores crearon la Encuesta Permanente de Empleo (EPE) 
- En un inicio la EPE era una encuesta panel fijo, es decir, los encuestadores retornaban cada trimestre a los mismos hogares.
<cite data-cite="undefined"></cite>

## Importancia

- Permite captar información de la oferta de mano de obra, es decir, de los trabajadores, en todas las categorias ocupacionales. 
- No sólo indaga sobre las características de las personas ocupadas sino de las que están buscando trabajo.  
- Genera indicadores laborales (empleo e ingresos) con reportes mensuales, con información de periodicidad trimestral (trimestre movil) a nivel de Lima Metropolitana <cite data-cite="undefined"></cite>

## Ficha técnica

#### Objetivos

- Genera indicadores de empleo e ingreso a nivel de LM
- Servir de fuente de información a instituciones públicas y privadas
- Permitir la comparabilidad con otras investigaciones 



**Principales indicadores**

- Magnitud del empleo (# de ocupados, horas trabajadas)
- Desempleo
- Subempleo
- Cruces por ramas de actividad, tamaño de empresa, sexo, grupo de edad, informalidad, etc. 
- Promedio de ingresos 


**Unidad de Analisis** 
- Integrantes del hogar familiar, trabajadores cama adentro y pensionistas máx. 9 pensionistas. 

**Diseño muestral** 
- **_tipo de muestreo_** es probabilistco, bietapico, estratificado,  
- **_muestra panel_** panel de rotacion de hogares 1/6 de la muestra cada trimestre.   



**Tamaño de muestra**
- Anual: 19200 viviendas. Trimestral Agosto-Setiembre-Octubre 2021 : 4800 viviendas. 

**Tipo de encuesta**
- Encuesta de derecho

**Informantes**
- Residentes habituales del hogar de 14 años a más. 

**Inferencia**
- Trimestral y a nivel del área de Lima Metropolitana
